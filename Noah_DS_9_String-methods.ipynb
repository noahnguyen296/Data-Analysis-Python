## 4. Splitting strings with `.split`

* The `.split` method allows us to split/break the text data into parts.
* The main argument of the method is
1. `sep =`: the character served as the split point. We can use `Regex` for this.
2. `expand = True/False` if we want to generate new columns and assign the split parts into these. Doing this will generate a Data Frame with columns containing the split parts.
3. `n =` specifies the numbers of split. For example the data is 'Thor / Captain / America' and if we use `sep = '/'`, then by default we receive `[Thor, Captain, America]`; if we add `n = 1` then we receive `[Thor, Captain / America]`.

* The `.split` method executes from left to right. If we want to apply from right to left, then we call `.rsplit`.

```python
titanic['home.dest']

# This column contain data of the passenger home and their destination
```

```python
# Suppose that we want to split into 2 parts: before and after '/'

titanic['home.dest'].str.split('/')
```

```python
# Note the new structure within each entry of the Series

type(titanic['home.dest'].str.split('/')[1])
```

```python
# to access elements in a list we can use 

titanic['home.dest'].str.split('/')[1][0]
```

```python
# We can assign the split parts into columns instead of storing as a list
# But this is less useful if the original string data are not in a similar structure
# For example a data with one '/' will be split into 2 columns, while another with two '/' will be split into 3 columns

titanic['home.dest'].str.split('/', expand = True)
```

```python
# Add new columns of passenger home and destinations

titanic['home'] = titanic['home.dest'].str.split('/', expand = True)[0]
titanic['destination'] = titanic['home.dest'].str.split('/', expand = True)[1]
```

## 7. Check if the data `.startswith` or `.endswith` certain values

* The methods `.startswith` and `.endswith` work similar to `.contains` but they will find exactly the entries that begin or end with a conditional substring.
* For example when we call `.contains('New York')` then the output includes those with data such as 'East New York'; 'Brooklyn New York'... Hence we can use `.startswith('New York')` if we only want the data exactly as 'New York'.

```python
UFO_report[UFO_report.city.str.startswith('New York', na = False)]
```

```python
# If we concern that there may be entries of, says, 'New York ABC' then we can use AND

(
    UFO_report[
    (UFO_report.city.str.startswith('New York', na = False)) &
    (UFO_report.city.str.endswith('New York', na = False))
    ]
)
```

```python
# sanity check

(
    UFO_report[
    (UFO_report.city.str.startswith('New York', na = False)) &
    (UFO_report.city.str.endswith('New York', na = False))
    ]
).city.unique()
```
