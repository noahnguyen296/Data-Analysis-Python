## **6. Data types and NaN**

```python
titanic.info()
```

* Some of the columns such as age, fare should be in the data type of integers or floats, but are being stored as objects.
* This hinders the use of numerical methods such as vectorization.

### 6.1. `.astype`

```python
titanic.age.unique()

# we need a first check
```

* If the entries are all in numeric presentations but are stored as objects, then we can apply the method `.astype` to convert directly.
* However, there is some values `?` and we need to address it first.

```python
# One option that we have practiced is .replace

titanic['age'].replace(['?'], [None], inplace = True)
```

```python
# Now we can apply .astype
# Here we create a new column to compare

titanic.insert(5, 'age_float', titanic.age.astype(float))
```

```python
titanic[['age', 'age_float']]
```

```python
# Now we can apply numeric methods on the column

print(titanic.age_float.mean())
print(titanic.age_float.median())
print(titanic.age_float.max())
print(titanic.age_float.min())
```

### 6.2. `pd.to_numeric`

* `.pd.to_numeric()` takes in a Series/scalar, then convert into numeric datatypes, either float or int, depends on the input.
* Useful parameter to address invalid input: `errors =` `'raise'`(raise an exception); `'coerce'`(set as NaN); `'ignore'`(return the original).

```python
titanic.fare.unique()
```

* This column contains `?`. As demonstrated above, we can first `.replace` the value `?` with `[None]` then use `.astype`.
* We see that `pd.to_numeric` can handle such problem directly by the parameters `errors = `.

```python
pd.to_numeric(titanic.fare, errors = 'coerce').sort_values() # set as NaN for invalid values, i.e. ?
```

### 6.3. Working with `NaN`

```python
game
```

#### `.isna`

```python
# .isna() returns Boolean outputs of whether the entries are NaN

game.isna()
```

#### `.dropna`

```python
# To drop NaN, we can use .dropna()
# First let's see how it works for a Series

print('Original series: ')
print(game['assists'])

print('After removing NaN: ')
print(game['assists'].dropna())
```

Notes when applying `.dropna()` on a DataFrame:
1) By default the parameter `axis = 0`, meaning that it will drop the entire row. We change to `axis = 1` if we need to drop columns with NaN.
2) By default the parameter `how = any`, meaning that it will drop the entire row containing NaN values in any columns. We have 2 selections: 'any' and 'all' (only rows with all NaN values are dropped).
3) We can add `subset = [col names]` to select the conditional column.

```python
# e.g. default setting

game.dropna()
```

```python
# Suppose we only want to drop rows that have missing values in the column 'rebounds'

game.dropna(subset = ['rebounds'])
```

#### `.fillna`

Noteworthy parameters:
1. `value = `: the arguments to replace NaN.
2. `method = `: default `None`. We can use this to provide the method to replace NaN, with options including `'backfill'`/`'bfill'` (backward filling); `'pad'`; `'ffill'` (forward filling).

```python
# example 1: fill a Series 

game.name.fillna('Unknown', inplace = True) # for data without the name, change to Unknown
game
```

```python
# example 2: fill the entire Dataframe

game.fillna(0)
```

```python
# example 3: fill multiple columns with different values

game.fillna(
    {
        'league' : 'others',          # for unknown league data, set to others
        'points' : 0,                 # for unknown points, assuming 0
        'assists' : game.assists.min(), # for unknown assits, fill with the min value 
        'rebounds' : game.rebounds.mean(), # for unknown assits, fill with the mean value 
    }
)
```

```python
# example 4: using method

print('Before filling: ')
print(game.points)

print('After forward filling: ')
print(game.points.fillna(method = 'ffill'))
```
